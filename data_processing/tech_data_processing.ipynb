```python
import re

def split_arr(text : str):
    spl = re.findall(r"'(.*?)'", str(text))
    return spl
```

```python
import random

def processing(row):
    category = id_by_cate[split_arr(row["category"])[2]]
    try:
        description = split_arr(row["description"])[0]
    except:
        description = ""
    try:
        image = split_arr(row["imageURLHighRes"])[0]
    except:
        image = ""
    body = {
        "name" : str(row["title"]),
        "description" : description,
        "status" : "PUBLIC",
        "category" : category,
        "price" : random.randint(2, 200),
        "quantity" : 100,
        "image" : image
    }
    resp = requests.post(URL_CREATE_POST,
                        data = json.dumps(body),
                        headers = {
                            "Content-Type" : "application/json",
                            "Authorization" : "Bearer %s" % TOKEN
                        })
```

```python
data.apply(func=processing, axis=1)
```
